```python
from sklearn.metrics import silhouette_score as shs

def gridsearch(combinations, x):

    scores = []
    all_labels = []

    for i, (eps, num_samples) in enumerate(combinations):
        dbscan_model = DBSCAN(eps=eps, min_samples=num_samples).fit(x)
        labels = dbscan_model.labels_
        labels_set = set(labels)
        num_clusters = len(labels_set)
        if -1 in labels_set:
            num_clusters -= 1
        if num_clusters < 2:  # Ensure there are at least 2 clusters
            scores.append(-1)  # Append a default low score if only 1 cluster is found
            all_labels.append(labels)
            continue
        scores.append(shs(x, labels))
        all_labels.append(labels)
        #rint("at iteration:", i, "score", scores[-1], "Number of clusters:", num_clusters)
    best_index = np.argmax(scores)
    best_parameters = combinations[best_index]
    best_labels = all_labels[best_index]
    best_score = scores[best_index]
    return {'best_eps': best_parameters[0],
            'best_min_samples': best_parameters[1],
            'best_labels': best_labels,
            'best_Score': best_score}
```

```python
# Unique city names
cities = df_train['city_name'].unique()
```

```python


# Store best parameters for each city
best_parameters_per_city = {}

# Iterate over each city and perform grid search
for city in cities:
    city_coords = coords_with_city[coords_with_city['city_name'] == city][['delivery_location_latitude_rad', 'delivery_location_longitude_rad']].to_numpy()
    best_para = gridsearch(combinations, city_coords)
    best_parameters_per_city[city] = best_para
    print(f"Best parameters for {city}: {best_para}")

# Display the best parameters for each city
for city, params in best_parameters_per_city.items():
    print(f"City: {city}, Best Eps: {params['best_eps']}, Best Min Samples: {params['best_min_samples']}, Best Score: {params['best_Score']}")
```

```python
# Initialize a column to store cluster labels
coords_with_city['cluster'] = -1  # Default value for noise points

# Keep track of the next cluster ID
next_cluster_id = 0

# Apply DBSCAN for each city using the best parameters and assign unique cluster labels
for city, params in best_parameters_per_city.items():
    city_coords = coords_with_city[coords_with_city['city_name'] == city][['delivery_location_latitude_rad', 'delivery_location_longitude_rad']].to_numpy()
    dbscan_model = DBSCAN(eps=params['best_eps'], min_samples=params['best_min_samples']).fit(city_coords)
    cluster_labels = dbscan_model.labels_
    # Assign unique sequential cluster labels
    new_cluster_labels = np.where(cluster_labels != -1, cluster_labels + next_cluster_id, cluster_labels)
    coords_with_city.loc[coords_with_city['city_name'] == city, 'cluster'] = new_cluster_labels
    # Update the next cluster ID only if there are valid clusters
    if new_cluster_labels.max() != -1:
        next_cluster_id = new_cluster_labels.max() + 1
```

Using HDBSCAN to cluster the noise points identified by DBSCAN can be a good strategy:

Adaptability to Noise: HDBSCAN (Hierarchical Density-Based Spatial Clustering of Applications with Noise) is designed to handle noise more effectively than DBSCAN. It uses a hierarchical approach to clustering, which allows it to detect clusters of varying densities and separate noise points more accurately.

Handling Sparse Data: Given that we have 3-4% noise points, HDBSCAN can be particularly useful. It can identify clusters within this sparse dataset where DBSCAN might have failed due to its fixed density threshold.

Stability: HDBSCAN can provide more stable clustering results than DBSCAN, especially in datasets where clusters have different densities. It uses a concept of cluster stability to determine which clusters are meaningful.

```python
# Now, apply HDBSCAN to noise points
import hdbscan
from sklearn.preprocessing import StandardScaler

# Data points not clustered by DBSCAN (labeled as -1)
noise_points = coords_with_city[coords_with_city['cluster'] == -1]

if not noise_points.empty:
    # Extract unique city names from the noise points
    unique_cities = noise_points['city_name'].unique()

    # Process each city separately
    for city in unique_cities:
        city_coords = noise_points[noise_points['city_name'] == city][['delivery_location_latitude_rad', 'delivery_location_longitude_rad']]
        
        if not city_coords.empty:
            # Standardize the coordinates for clustering
            scaler = StandardScaler()
            city_coords_scaled = scaler.fit_transform(city_coords)
            
            # Apply HDBSCAN to the city's coordinates
            hdbscan_model = hdbscan.HDBSCAN(min_cluster_size=2)
            city_clusters = hdbscan_model.fit_predict(city_coords_scaled)
            
            # Ensure HDBSCAN labels are unique and sequential
            city_clusters = np.where(city_clusters != -1, city_clusters + next_cluster_id, city_clusters)
            
            # Assign new cluster labels to the city's data points
            coords_with_city.loc[(coords_with_city['city_name'] == city) & (coords_with_city['cluster'] == -1), 'cluster'] = city_clusters
            
            # Update next_cluster_id to ensure unique cluster IDs
            if city_clusters.max() != -1:
                next_cluster_id = city_clusters.max() + 1
```

```python
model_A = ['Agra', 'Aurangabad', 'Bhopal', 'Delhi', 'Goa', 'Kanpur', 'Kochi', 'Kolkata', 'Ludhiana','Prayagraj']
model_B = ['Bangalore', 'Chennai', 'Coimbatore','Hyderabad','Indore','Jaipur','Mumbai','Mysore','Pune','Ranchi','Surat','Vadodra']
```

```python
# Function to process each group
def process_group(group, time_bands):
    # Create a time series with count of orders per 15-minute interval
    order_counts = group.resample('15T').size()
    
    # Apply rolling window of 1 hour and sum the counts
    rolling_orders = order_counts.rolling('1H').sum()
    
    # Align the rolling orders with the common time bands, filling in any gaps with 0
    aligned_rolling_orders = rolling_orders.reindex(time_bands, fill_value=0)
    
    # Create a time band label for each window
    time_labels = [f"{time.strftime('%Y-%m-%d %H:%M:%S')} to {(time + pd.Timedelta(hours=1)).strftime('%Y-%m-%d %H:%M:%S')}" for time in aligned_rolling_orders.index]
    
    return pd.DataFrame({
        'time_band': time_labels,
        'hourly_orders': aligned_rolling_orders.values
    })

# Function to get the min and max time bands for a city group
def get_time_bands_for_group(city_group):
    min_time = city_group.index.min()
    max_time = city_group.index.max()
    return pd.date_range(start=min_time, end=max_time, freq='15T')

# Process each city and cluster group and combine the results
aggregated_df_list_model_A = []
aggregated_df_list_model_B = []

# Process Model A cities
model_A_cities = df_train[df_train['city_name'].isin(model_A)]
model_A_time_bands = get_time_bands_for_group(model_A_cities)

for (city, cluster), group in model_A_cities.groupby(['city_name', 'cluster']):
    processed_group = process_group(group, model_A_time_bands)
    processed_group['city_name'] = city
    processed_group['cluster'] = cluster
    processed_group['model_type'] = 'Model_A'
    aggregated_df_list_model_A.append(processed_group)

# Process Model B cities
model_B_cities = df_train[df_train['city_name'].isin(model_B)]
model_B_time_bands = get_time_bands_for_group(model_B_cities)

for (city, cluster), group in model_B_cities.groupby(['city_name', 'cluster']):
    processed_group = process_group(group, model_B_time_bands)
    processed_group['city_name'] = city
    processed_group['cluster'] = cluster
    processed_group['model_type'] = 'Model_B'
    aggregated_df_list_model_B.append(processed_group)

# Concatenate all processed groups into separate dataframes for each model
aggregated_df_model_A = pd.concat(aggregated_df_list_model_A).reset_index(drop=True)
aggregated_df_model_B = pd.concat(aggregated_df_list_model_B).reset_index(drop=True)
```

```python

# Function to categorize each timestamp into one of the one-hour time bands
def get_time_band(timestamp):
    hour = timestamp.hour
    return f"{hour:02d}:00 to {(hour+1)%24:02d}:00"

# Apply the function to create a new 'time_band' column
df_train['time_band_only_hours'] = df_train.index.to_series().apply(get_time_band)

# Group the data by date and time band
grouped = df_train.groupby(['order_date', 'time_band_only_hours'])['weather_conditions'].apply(lambda x: x.mode().iloc[0] if not x.mode().empty else None).reset_index()

# Pivot the table to get time bands as columns
pivot = grouped.pivot(index='order_date', columns='time_band_only_hours', values='weather_conditions')

# Fill NaN values with a placeholder indicating no data
pivot = pivot.fillna('No Data')

# Ensure that all expected columns are present in the pivot DataFrame
time_band_columns = [f"{hour:02d}:00 to {(hour+1)%24:02d}:00" for hour in range(24)]

for column in time_band_columns:
    if column not in pivot.columns:
        pivot[column] = 'No Data'

# Calculate the time difference between changes for each pair of consecutive time bands
for i in range(len(time_band_columns) - 1):
    pivot[time_band_columns[i]] = (pivot[time_band_columns[i]] != pivot[time_band_columns[i + 1]]).astype(int) * 1

# Handle the wrap-around case
pivot[time_band_columns[-1]] = (pivot[time_band_columns[-1]] != pivot[time_band_columns[0]]).astype(int) * 1

# Calculate the total hours of change for each day
pivot['total_change_hours'] = pivot[time_band_columns].sum(axis=1)

# Calculate the average number of hours in which the weather changes
avg_change_hours = pivot['total_change_hours'].mean()

print(f"Average number of hours in which the weather changes: {avg_change_hours}")
```

```python
# Extract the start and end time from time_band and convert to datetime
for df in [aggregated_df_model_A, aggregated_df_model_B]:
    df[['start_time', 'end_time']] = df['time_band'].str.split(' to ', expand=True)
    df['start_time'] = pd.to_datetime(df['start_time'])
    df['end_time'] = pd.to_datetime(df['end_time'])

# Step 1: Resample df_train to 12-hour intervals and compute mode
def compute_mode(series):
    try:
        return series.mode().iloc[0]
    except IndexError:  # If no mode found
        return "NA"

weather_12h = df_train.groupby('city_name').resample('12H').agg({'weather_conditions': compute_mode})


# Flatten the index and rename columns for merging
weather_12h.reset_index(inplace=True)
weather_12h.rename(columns={'order_timestamp': '12h_band_start_time'}, inplace=True)
weather_12h['12h_band_end_time'] = weather_12h['12h_band_start_time'] + pd.Timedelta(hours=12)

# Check if the time bands of aggregated_df_model_A and B fall within the 12-hour bands
def map_weather(row):
    start_time = row['start_time']
    end_time = row['end_time']
    city = row['city_name']
    
    weather_conditions = weather_12h[
        (weather_12h['city_name'] == city) &
        (weather_12h['12h_band_start_time'] <= start_time) &
        (weather_12h['12h_band_end_time'] >= end_time)
    ]['weather_conditions']
    
    return weather_conditions.iloc[0] if not weather_conditions.empty else "NA"

for df in [aggregated_df_model_A, aggregated_df_model_B]:
    df['weather_conditions'] = df.apply(map_weather, axis=1)

    # Drop unnecessary columns
    df.drop(columns=['start_time', 'end_time'], inplace=True)
```

```python

# Function to replace 'NA' string, None, and NaN values with the next valid value or preceding value if needed
def fill_valid_value(series):
    mask_na = series.isin(['NA', None]) | pd.isnull(series)
    
    # Fill forward
    for i in range(len(series) - 2, -1, -1):
        if mask_na[i]:
            for j in range(i + 1, len(series)):
                if series[j] not in ['NA', None] and not pd.isnull(series[j]):
                    series[i] = series[j]
                    break
    
    # Fill backward
    for i in range(1, len(series)):
        if mask_na[i]:
            for j in range(i - 1, -1, -1):
                if series[j] not in ['NA', None] and not pd.isnull(series[j]):
                    series[i] = series[j]
                    break
    
    return series

# Function to process the DataFrame in chunks
def process_in_chunks(df, chunk_size, columns):
    chunks = []
    df = df.reset_index(drop=True)  # Reset the index to ensure it is sequential
    total_rows = len(df)
    print(f"Total rows: {total_rows}")

    for start in range(0, total_rows, chunk_size):
        end = min(start + chunk_size, total_rows)  # Ensure the end index does not exceed the DataFrame length
        print(f"Processing rows {start} to {end}")
        chunk = df.iloc[start:end].copy()
        for col in columns:
            print(f"Processing column: {col}")
            try:
                chunk[col] = fill_valid_value(chunk[col])
            except KeyError as e:
                print(f"KeyError: {e} for column {col} in chunk {start} to {end}")
                continue
        chunks.append(chunk)
    
    result = pd.concat(chunks, axis=0).reset_index(drop=True)
    return result

# Apply the function to the specified columns for both dataframes
columns_to_fill = ['weather_conditions_lag1', 'weather_conditions_lag2', 'hourly_orders_lag1', 'hourly_orders_lag5']
chunk_size = 500  # Adjust chunk size based on your system's memory capacity

# Apply chunk processing
final_df = process_in_chunks(final_df, chunk_size, columns_to_fill)

```

```python
import optuna
from optuna.integration import XGBoostPruningCallback
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error
import pandas as pd
import numpy as np


def objective(trial):
    # Define a sequential split point for the validation set
    inner_split_point = int(len(X_train) * 0.80)
    
    # Further split the training data into training and validation sets sequentially
    X_train_inner = X_train.iloc[:inner_split_point]
    X_valid = X_train.iloc[inner_split_point:]
    y_train_inner = y_train.iloc[:inner_split_point]
    y_valid = y_train.iloc[inner_split_point:]

    param_grid = {
        'n_estimators': trial.suggest_int('n_estimators', 100, 1000),
        'max_depth': trial.suggest_int('max_depth', 3, 10),
        'min_child_weight': trial.suggest_int('min_child_weight', 3, 10),
        'learning_rate': trial.suggest_float('learning_rate', 0.05, 0.4, log=True),
        'gamma': trial.suggest_float('gamma', 1, 10, log=True),
        'subsample': trial.suggest_float('subsample', 0.5, 1),
        'colsample_bytree': trial.suggest_float('colsample_bytree', 0.5, 1),
        'reg_alpha': trial.suggest_float('reg_alpha', 0.01, 1, log=True),
        'reg_lambda': trial.suggest_float('reg_lambda', 0.01, 1, log=True)
    }
    xgb_optuna_model = XGBRegressor(**param_grid, random_state=42)

    # Fit the model with the training and validation data
    xgb_optuna_model.fit(X_train_inner, y_train_inner, eval_set=[(X_valid, y_valid)], verbose=False)

    # Make predictions and calculate loss
    y_valid_pred = xgb_optuna_model.predict(X_valid)
    loss = mean_squared_error(y_valid, y_valid_pred)
    return loss

study = optuna.create_study(direction='minimize')  # Minimize in case of mean squared error
study.optimize(objective, n_trials=500)
best_params = study.best_params
print(f"Best parameters: {best_params}")

```

```python
import pandas as pd
import numpy as np
import xgboost as xgb
from sklearn.preprocessing import OneHotEncoder
from datetime import datetime, timedelta

# List of cities
cities = [
    'Jaipur', 'Ranchi', 'Bangalore', 'Surat', 'Hyderabad', 'Mumbai', 'Mysore',
    'Coimbatore', 'Vadodra', 'Indore', 'Chennai', 'Pune', 'Agra', 'Ludhiana',
    'Prayagraj', 'Kanpur', 'Delhi', 'Goa', 'Aurangabad', 'Kochi', 'Kolkata', 'Bhopal'
]

# One-hot encode the city_name column using the list of cities
ohe = OneHotEncoder(sparse=False)
ohe.fit(pd.DataFrame(cities, columns=['city_name']))

# Define your feature names based on your training data
feature_names = [
    f'cluster_{i}' for i in range(595)
] + [
    'festival', 'is_weekend', 'hourly_orders_lag1', 'hourly_orders_lag5', 
    'weather_conditions_lag1_Fog', 'weather_conditions_lag1_Sandstorms', 'weather_conditions_lag1_Stormy', 
    'weather_conditions_lag1_Sunny', 'weather_conditions_lag1_Windy',
    'weather_conditions_lag2_Fog', 'weather_conditions_lag2_Sandstorms', 'weather_conditions_lag2_Stormy', 
    'weather_conditions_lag2_Sunny', 'weather_conditions_lag2_Windy',
    'start_year', 'start_month', 'start_day', 'start_hour', 'start_minute', 
    'end_year', 'end_month', 'end_day', 'end_hour', 'end_minute',
    'city_name_Agra', 'city_name_Aurangabad', 'city_name_Bangalore', 'city_name_Bhopal', 
    'city_name_Chennai', 'city_name_Coimbatore', 'city_name_Delhi', 'city_name_Goa', 
    'city_name_Hyderabad', 'city_name_Indore', 'city_name_Jaipur', 'city_name_Kanpur', 
    'city_name_Kochi', 'city_name_Kolkata', 'city_name_Ludhiana', 'city_name_Mumbai', 
    'city_name_Mysore', 'city_name_Prayagraj', 'city_name_Pune', 'city_name_Ranchi', 
    'city_name_Surat', 'city_name_Vadodra'
]

# Function to prepare new data for prediction
def prepare_new_data(city, start_year, start_month, start_day, start_hour, start_minute, is_weekend, festival, clusters, weather_conditions_lag1, weather_conditions_lag2, hourly_orders_lag1, hourly_orders_lag5):
    # One-hot encode city
    city_encoded = ohe.transform([[city]])
    city_encoded_df = pd.DataFrame(city_encoded, columns=ohe.get_feature_names_out(['city_name']))

    # One-hot encode weather conditions lag1 and lag2
    weather_conditions = ['Fog', 'Sandstorms', 'Stormy', 'Sunny', 'Windy']
    weather_lag1_encoded = [1 if wc == weather_conditions_lag1 else 0 for wc in weather_conditions]
    weather_lag2_encoded = [1 if wc == weather_conditions_lag2 else 0 for wc in weather_conditions]

    # Prepare cluster data
    cluster_data = np.zeros(595)  # Assuming 595 clusters
    for cluster in clusters:
        cluster_data[cluster] = 1

    # Calculate end time based on a fixed 1-hour time band
    start_time = datetime(start_year, start_month, start_day, start_hour, start_minute)
    end_time = start_time + timedelta(hours=1)
    end_year = end_time.year
    end_month = end_time.month
    end_day = end_time.day
    end_hour = end_time.hour
    end_minute = end_time.minute

    # Combine all features
    new_data = np.concatenate([
        cluster_data,
        [festival, is_weekend, hourly_orders_lag1, hourly_orders_lag5],
        weather_lag1_encoded,
        weather_lag2_encoded,
        [start_year, start_month, start_day, start_hour, start_minute, end_year, end_month, end_day, end_hour, end_minute],
        city_encoded_df.values.flatten()
    ])
 
    new_data_df = pd.DataFrame([new_data], columns=feature_names)

    return new_data_df

# Function to make a prediction
def predict_new_data(model, city, start_year, start_month, start_day, start_hour, start_minute, is_weekend, festival, clusters, weather_conditions_lag1, weather_conditions_lag2, hourly_orders_lag1, hourly_orders_lag5):
    new_data = prepare_new_data(city, start_year, start_month, start_day, start_hour, start_minute, is_weekend, festival, clusters, weather_conditions_lag1, weather_conditions_lag2, hourly_orders_lag1, hourly_orders_lag5)
    
  
    
    prediction = model.predict(new_data)
    return prediction

```
